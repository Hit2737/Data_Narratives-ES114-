### **Dataset 1**

```python
import pandas as pd
import numpy as np

AOm = pd.read_csv('AusOpen-men-2013.csv')
AOm = AOm.replace(np.NaN,0)
AOm
```

**Question1:** Compare how the player 1 behaves on his increasing set score and his oppenent's incresing set score.*italicized text*

```python
import numpy as np
import matplotlib.pyplot as plt

Set_Results1 = AOm[['ST1.1','ST2.1','ST3.1','ST4.1','ST5.1']]
Set_Results1.loc[:, 'Avg_set_result1'] = np.mean(Set_Results1,axis=1)
Set_Results2 = AOm[['ST1.2','ST2.2','ST3.2','ST4.2','ST5.2']]
Set_Results2.loc[:, 'Avg_set_result2'] = np.mean(Set_Results2,axis=1)

aces_by_set_result1 = AOm.groupby(Set_Results1['Avg_set_result1'])['ACE.1'].mean().reset_index()
aces_by_set_result2 = AOm.groupby(Set_Results2['Avg_set_result2'])['ACE.2'].mean().reset_index()
  
plt.figure(1,figsize=(10,5))
plt.bar(x='Avg_set_result1', height='ACE.1', data=aces_by_set_result1,width=0.15, color='blue', edgecolor='black')
plt.bar(x='Avg_set_result2', height='ACE.2', data=aces_by_set_result2, width=0.15, color='red', edgecolor='black')
plt.xlabel('Average Set Result of Player 1 and 2')
plt.ylabel('Average Aces by Player 1')
plt.legend(['Average Set Result of Player 1','Average Set Result of Player 2'])
plt.title('Average Number of Aces by Player 1 vs Avg. Set Results of Player 1')
plt.xticks()
plt.show()
```

### **Dataset 5**

```python
import pandas as pd
import numpy as np

USOm = pd.read_csv('USOpen-men-2013.csv')
USOm = USOm.replace(np.NaN,0)
USOm
```

**Question5:** Find an linear approximation of the Total points won based on the First serve wins by player 1 using Linear Regression and plot the respective graph.

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

X,Y_train = USOm[['FSW.1']],USOm[['TPW.1']]
LR = LinearRegression()
LR.fit(X.values,Y_train)
Y_predicted = LR.predict(X.values)
plt.scatter(X,Y_train)
# For finding the equation of the approximated line:
y_0 = LR.predict([[0]])   # y_0 = m x 0 + c => y_0 = c
y_1 = LR.predict([[1]])   # y_1 = m x 1 + y_0  => m = y_1 - y_0
m = y_1 - y_0
print(f"The equation of approximated line is:\033[1;43m y = {m[0,0]:.2f}x + {y_0[0,0]:.2f} ")
plt.plot(X,Y_predicted,'r-')
plt.title('Total points win VS First serve wins')
plt.xlabel('First Serves won by Player 1');plt.ylabel('Total Points win by Player 1')
plt.legend(["Dataset points","y=2.02x + 17.24"])
plt.show()
```

**Question7:** Find the player which has least unforced errors and then calculate the number of Winner titles he has won?

```python
Df = np.argmin(Wm['UFE.1']),np.argmin(Wm['UFE.2'])
idx = np.min(Df)
if (Wm.loc[idx,'UFE.1']<Wm.loc[idx,'UFE.2']):
  print(f"The player having least unforced errors is {Wm.loc[idx,'Player1']} with only {Wm.loc[idx,'UFE.1']} unforced errors.")
  print(f"The number of Winner titles he has won are {Wm.loc[idx,'WNR.1']}")
else:
  print(f"The player having least unforced errors is {Wm.loc[idx,'Player2']} with only {Wm.loc[idx,'UFE.2']} unforced errors.")
  print(f"The number of Winner titles he has won are {Wm.loc[idx,'WNR.2']}")

print("This implies that less the unforced errors happen to you more you win.")
```

```python
import pandas as pd
import numpy as np
Ww = pd.read_csv('Wimbledon-women-2013.csv')
Ww = Ww.replace(np.NaN,0)
Ww
```

**Question8:** What is the accuracy percentage of winning net points by player 1? Also plot the percentage graph for player 1?

```python
import matplotlib.pyplot as plt
import seaborn as sns

dF = Ww[['NPA.1','NPW.1']]
dF.loc[:,'%NPW/NPA1'] = dF.loc[:,'NPW.1']/dF.loc[:,'NPA.1']*100

xaxis = np.arange(len(dF['%NPW/NPA1']))
plt.figure(figsize = (20,7))
sns.stripplot(x=xaxis, y = '%NPW/NPA1', data = dF , hue ='%NPW/NPA1' )
for i in range(len(dF['%NPW/NPA1'])):
    plt.plot([xaxis[i], xaxis[i]], [0, dF['%NPW/NPA1'][i]], linestyle='--', color='gray')
plt.xticks(rotation = 90)
plt.xlabel('Player 1 Index');plt.ylabel('Accuarcy Percentage in scoring Net Points')
plt.show()
```
